```python
from selenium import webdriver
from time import sleep
from selenium.webdriver.chrome.service import Service
import pandas as pd

# 1. Khai báo biến browser
# Đường dẫn đến driver
chrome_driver_path = "D:/TuHocAI/DuDoanKetQuaXS/chromedriver-win64/chromedriver.exe"
# Khởi tạo Chrome Driver Service
service = Service(executable_path=chrome_driver_path)

# Khởi tạo trình duyệt Chrome
browser = webdriver.Chrome(service=service)


# # 2. Mở thử 1 trang web
# browser.get("https://www.thantai1.net/so-ket-qua")
```

```python
from datetime import datetime, timedelta
current_date = datetime(2024,9,9)
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
data = []
#Read utl
idx = 0
while True:
    print("Process 300 days from {}-{}-{}".format(current_date.day, current_date.month, current_date.year))
    url = "https://www.thantai1.net/so-ket-qua"
    browser.get(url)

    #Set date
   # end = browser.find_element_by_id("end")
    wait = WebDriverWait(browser, 10)
    # Tìm phần tử end (set ngày)
    end = wait.until(EC.presence_of_element_located((By.ID, "end")))
   # end = browser.find_element("id", "end")
    end.clear()
    end.send_keys("{}-{}-{}".format(current_date.day, current_date.month, current_date.year))
    
    #Nhấn nút tìm kiếm
    btn = browser.find_element("xpath", "/html/body/div[3]/main/div/form/div[2]/div/button[9]")
    btn.click()

   # Tìm kết quả xổ số (dùng css selector với nhiều class)
    result = wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, ".font-weight-bold.text-danger.col-12.d-block.p-1.m-0")))

    for row in result:
        print(row.text)
        idx = idx+1
        data.append(row.text)
    # Giảm ngày đi 300 ngày
    current_date -= timedelta(days=300)
    
    # Dừng khi đủ slg kết quả
    if idx > 20*365:
        break

df = pd.DataFrame(data, columns=['KQ'])
df.to_csv("XSMB.csv", index = False)
browser.quit()
```
